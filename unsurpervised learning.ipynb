```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.cluster import KMeans,hierarchical
from sklearn.metrics import accuracy_score,cohen_kappa_score,confusion_matrix,adjusted_rand_score,silhouette_score
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
i=load_iris()
```

```python
#Iris data in the form of array
i.data
```

```python
#Now creating the dataframe
iris=pd.DataFrame(i.data,columns=i.feature_names)
```

```python
iris.head()
```

```python
i.target
```

```python
#initilazing k means where number of clusters are 3.(given)
k=KMeans(n_clusters=3)
labels=k.fit_predict(iris)
```

```python
labels
```

```python
k.inertia_
```

```python
k.cluster_centers_
```

```python
iris.columns
```

```python
# Now using the two features such as sepal length and sepal width
#We can also plot the cluster centres as centroids between the scatter to have rep. of centroids. 
plt.scatter(iris["sepal length (cm)"],iris["sepal width (cm)"],c=labels)
plt.scatter(k.cluster_centers_[:,0],k.cluster_centers_[:,1],marker="d",c="red",s=70)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal width (cm)")
```
